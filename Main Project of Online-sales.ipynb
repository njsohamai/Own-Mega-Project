```python
import tkinter as tk
import customtkinter as ctk
import pandas as pd
from datetime import datetime
import os
from PIL import Image, ImageTk

def place_order():
    name = name_entry.get()
    age = age_entry.get()
    gender = gender_entry.get()
    number = number_entry.get()
    address = address_entry.get()
    product = product_entry.get()
    stock_code = stock_code_entry.get()
    category = category_entry.get()
    quantity = quantity_entry.get()
    price = float(price_entry.get())
    invoiceno = invoice_entry.get()
    hsn = hsn_entry.get()
    mode = mode_entry.get()

    filename = "Online-Sales Data.xlsx"

    if not all([name,age,gender,number,address,product,stock_code,category,quantity,price,invoiceno,hsn,mode]):
        tk.messagebox.showwarning('Input Error','Please fill all the details')
        return

    # Calculate Gst, Shipping Cost and Discount
    gst = price * 0.18
    shipping = price * 0.025
    discount = price * 0.02

    curr_date = datetime.now().strftime('%d-%m-%y %H:%M:%S')
    if filename in os.listdir():
        df=pd.read_excel(filename)
        df.loc[len(df)]=[name,age,gender,number,address,product,stock_code,category,quantity,price,discount,shipping,gst,invoiceno,hsn,mode,curr_date]
        df.to_excel('Online-Sales Data.xlsx',index=False)
    else:
        # Prepare data to excel
        data = {
            "Name":[name],
            "Age":[age],
            "Gender":[gender],
            "Phone Number":[number],
            "Address":[address],
            "Product Name":[product],
            "Stock Code":[stock_code],
            "Product Category":[category],
            "Quantity Sold":[quantity],
            "Product Price":[price],
            "Discount Given":[discount],
            "Shipping Cost":[shipping],
            "Gst Amount":[gst],
            "Invoice Number":[invoiceno],
            "HSN Code":[hsn],
            "Payment Mode":[mode],
            "Date Time":[curr_date]
        }
        df = pd.DataFrame(data)
        df.to_excel('Online-Sales Data.xlsx', index=False)
    tk.messagebox.showinfo('Sucess', 'Your Order is Placed')

# Creating a window
window = ctk.CTk()
window.title('Online-sales Shooping Software')
window.geometry("600x600")

# Adjust the apparance
ctk.set_appearance_mode("Dark")
ctk.set_default_color_theme("blue")

#Create Title Label
title_label = ctk.CTkLabel(window,text="Online-Sales Customer Input", font=('Arial',45,"bold"))
title_label.place(relx=0.30,rely=0.05)

#Create Input Labels
name_label = ctk.CTkLabel(window, text='Customer Name:', font=('Arial',15,'bold'))
name_label.place(relx=0.25,rely=0.25,anchor='e')
name_entry = ctk.CTkEntry(window)
name_entry.place(relx=0.255,rely=0.25,anchor='w')

age_label = ctk.CTkLabel(window, text='Customer Age:', font=('Arial',15,'bold'))
age_label.place(relx=0.25,rely=0.35,anchor='e')
age_entry = ctk.CTkEntry(window)
age_entry.place(relx=0.255,rely=0.35,anchor='w')

gender_label = ctk.CTkLabel(window, text='Customer Gender:', font=('Arial',15,'bold'))
gender_label.place(relx=0.25,rely=0.45,anchor='e')
gender_entry = ctk.CTkEntry(window)
gender_entry.place(relx=0.255,rely=0.45,anchor='w')

number_label = ctk.CTkLabel(window, text='Customer Phone.No:', font=('Arial',15,'bold'))
number_label.place(relx=0.25,rely=0.55,anchor='e')
number_entry = ctk.CTkEntry(window)
number_entry.place(relx=0.255,rely=0.55,anchor='w')

address_label = ctk.CTkLabel(window, text='Customer Adress:', font=('Arial',15,'bold'))
address_label.place(relx=0.25,rely=0.65,anchor='e')
address_entry = ctk.CTkEntry(window)
address_entry.place(relx=0.255,rely=0.65,anchor='w')

product_label = ctk.CTkLabel(window, text='Product Name:', font=('Arial',15,'bold'))
product_label.place(relx=0.25,rely=0.75,anchor='e')
product_entry = ctk.CTkEntry(window)
product_entry.place(relx=0.255,rely=0.75,anchor='w')

stock_code_label = ctk.CTkLabel(window, text='Product Code:', font=('Arial',15,'bold'))
stock_code_label.place(relx=0.25,rely=0.85,anchor='e')
stock_code_entry = ctk.CTkEntry(window)
stock_code_entry.place(relx=0.255,rely=0.85,anchor='w')

category_label = ctk.CTkLabel(window, text='Product Category:', font=('Arial',15,'bold'))
category_label.place(relx=0.25,rely=0.95,anchor='e')
category_entry = ctk.CTkEntry(window)
category_entry.place(relx=0.255,rely=0.95,anchor='w')

quantity_label = ctk.CTkLabel(window, text='Quantity Purchase:', font=('Arial',15,'bold'))
quantity_label.place(relx=0.55,rely=0.25,anchor='e')
quantity_entry = ctk.CTkEntry(window)
quantity_entry.place(relx=0.555,rely=0.25,anchor='w')

price_label = ctk.CTkLabel(window, text='Product Price:', font=('Arial',15,'bold'))
price_label.place(relx=0.55,rely=0.35,anchor='e')
price_entry = ctk.CTkEntry(window)
price_entry.place(relx=0.555,rely=0.35,anchor='w')

invoice_label = ctk.CTkLabel(window, text='Invoice No:', font=('Arial',15,'bold'))
invoice_label.place(relx=0.55,rely=0.45,anchor='e')
invoice_entry = ctk.CTkEntry(window)
invoice_entry.place(relx=0.555,rely=0.45,anchor='w')

hsn_label = ctk.CTkLabel(window, text='HSN Code:', font=('Arial',15,'bold'))
hsn_label.place(relx=0.55,rely=0.55,anchor='e')
hsn_entry = ctk.CTkEntry(window)
hsn_entry.place(relx=0.555,rely=0.55,anchor='w')

mode_label = ctk.CTkLabel(window, text='Mode of Payment:', font=('Arial',15,'bold'))
mode_label.place(relx=0.55,rely=0.65,anchor='e')
mode_entry = ctk.CTkEntry(window)
mode_entry.place(relx=0.555,rely=0.65,anchor='w')

submit_button = ctk.CTkButton(window,text="Place Order",command=place_order,width=180,height=160)
submit_button.place(relx=0.75,rely=0.45)

img_path=os.getcwd()+'\\logo.jpg'

pil_image=Image.open(img_path)
logo_img=ctk.CTkImage(light_image=pil_image,size=(100,100))
logo_label=ctk.CTkLabel(window,text='',image=logo_img)
logo_label.place(relx=0.20, rely=0.001)

window.state('zoomed')

window.resizable(True,False)
window.geometry("2024x2024px")
window.mainloop()
```
